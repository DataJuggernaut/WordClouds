```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
from PIL import Image # converting images into arrays
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches # needed for waffle Charts
mpl.style.use('ggplot') # optional: for ggplot-like style
# check for latest version of Matplotlib
print ('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

```python
#Import World Cloud Modules
from wordcloud import WordCloud, STOPWORDS
print ('Wordcloud is installed and imported!')
```

## CHANGE TEXT FILE ADDRESS

```python
TextFile = open('C://Users/jalma/Documents/Data Science/IBM Data Science Certification/Word Clouds/AllFour.txt', 'r', encoding='utf-8').read()
print ('File downloaded and saved!')
```

```python
stopwords = set(STOPWORDS)
```

## GENERATE WORDCLOUD WITH TEXT FILE

```python
# instantiate a word cloud object
wordcloud_object = WordCloud(
    background_color='black',
    max_words=8000,
    stopwords=stopwords
)

# generate the word cloud
wordcloud_object.generate(TextFile)
```

## GENERATE WORDCLOUD AND SEARCH FOR MORE STOPWORD, WORDS UNEEDED

```python
wordcloud_object.generate(TextFile)
fig = plt.figure()
fig.set_figwidth(18) # set width
fig.set_figheight(22) # set height

# display the cloud
plt.imshow(wordcloud_object, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## ADD STOPWORDS HERE AND REGENERATE WORDCLOUD, INSPECT FOR ANY OTHER STOPWORDS

```python
stopwords.add('talk') # add the words said to stopwords
stopwords.add('say')
stopwords.add('said')
stopwords.add('will')
stopwords.add('be')
stopwords.add('the year')
stopwords.add('across the')
stopwords.add('across')
stopwords.add('called')
stopwords.add('of')
stopwords.add('began')
stopwords.add('in')
stopwords.add('of George')
stopwords.add('asked')
stopwords.add('of')
stopwords.add('seen')
stopwords.add('one')
stopwords.add('day')
stopwords.add('year')
stopwords.add('Friday')
stopwords.add('added')
stopwords.add('including')
stopwords.add('think')
stopwords.add('We')
stopwords.add('WE')

# re-generate the word cloud
wordcloud_object.generate(TextFile)

fig = plt.figure()
fig.set_figwidth(18) # set width
fig.set_figheight(22) # set height

# display the cloud
plt.axis('off')
plt.grid(None)
plt.imshow(wordcloud_object, interpolation='bilinear')
plt.axis('off')
plt.show()
```
